```python
#!/usr/bin/python3

from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
```

```python
hdulist = fits.open('hydrogen.fits')
#print(hdulist.info())
hdu = hdulist[0]
print(hdu.header)
```

```python
# Plot the 2D array
plt.imshow(hdulist)
plt.xlabel('x-pixels (RA)')
plt.ylabel('y-pixels (Dec)')
plt.colorbar()
plt.show()
```
